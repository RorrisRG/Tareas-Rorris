1. El perímetro de un triángulo isósceles es de $10 cm$. ¿Cuánto deben medir sus lados para que el volumen del cuerpo generado por la rotación del triángulo en torno a su base sea el mayor posible? (Volumen de un cono $= \frac{1}{3}\pi r^2 h$, donde $r$ es el radio de la base y $h$ la altura del cono).

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
l = sym.var('l', positive=True)
b = 10-2*l
h = sym.sqrt(l**2-(b/2)**2)
r = b/2
V = sym.pi*r**2*h/3
```

#Importamos librerias y definimos las variables l, b, h, r y V. Como queremos que todo quede respecto a $l$ definimos la base como $10-2l$. ($l$ siendo los lados iguales)

```python
dV = sym.diff(V, l)
c = sym.solve(sym.simplify(dV), )
display(V, dV, c, sym.N(V.subs(l, c[0])))
```

#Derivamos y sacamos los puntos críticos (en este caso marca a 5 como un mínimo por lo cual lo descartamos), la solución la encontramos con $V = 9.366...m^3$ cuando $l = 3$ al evaluar V(3) en (sym.N(V.subs(lados, c[$0$])))

```python
x = np.linspace(2.5, 5, 400)
f = sym.lambdify(l, V, 'numpy')
y = f(x)
plt.plot(x, y)
plt.plot(float(c[0]), f(float(c[0])), "ro")
plt.grid()
```

#Graficamos para visualizar la curva de la función y el máximo en 3.

2. Disponemos de una barra de aluminio de 6 metros para construir una portería de fútbol. Si queremos que el área de la portería sea máxima, ¿cuánto deben medir los postes y el travesaño?

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
L = sym.var('L')
def A(L):
    return L*(6-2*L)
```

#Definimos la función de área de la porteria en función de los postes L, sabiendo que A = 6 = 2l+L

```python
dL = sym.diff(A(L), L)
c = sym.solve(dL, L)
dL, c, A(1.5)
```

#Derivamos y conseguimos el máximo, notando que está en L = 3/2

```python
x = np.linspace(0, 3, 400)
y = A(x)
plt.plot(x, y)
plt.plot(float(c[0]), A(float(c[0])), "ro")
plt.grid()
```

#El máximo se encuentra en 1.5, por lo que el travesaño tendría que medir 3 metros y así cuando el travesaño mida 3 metros y los postes 1.5 metros conseguiremos un área máxima de $4.5m^2$
