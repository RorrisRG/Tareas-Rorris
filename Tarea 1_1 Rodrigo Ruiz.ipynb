# Optimización escalar con SymPy: criterio conocido

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

Para la siguientes funciones encuentre los puntos críticos, grafique y diga si éste se trata de un máximo o mínimo, local o global.

1. $f(x)=x+\frac{1}{x^2}$

```python
x = sym.var('x')
def f(x):
  return x + (1/(x**2))

df = sym.diff(f(x))
d2f = sym.diff(df)
c = sym.solve(df, x)
display(df, d2f, c)
x = np.linspace(-3, 3, 400)
y = f(x)
plt.plot(x, y)
plt.plot(c[0], f(c[0]), "ro")
plt.grid()
```

#Al tener la indeterminación en 0 no tenemos minimos ni maximos, solo asintotas, el único mínimo local real se encuentra en x = 1.259921

2. $f(x)=1260+117x-9x^2$

```python
x = sym.var('x')
def f(x):
  return 1200+117*x-9*x**2

df = sym.diff(f(x))
d2f = sym.diff(df)
c = sym.solve(df, x)
display(df, d2f, float(c[0]))
x = np.linspace(-3, 16, 400)
y = f(x)
plt.plot(x, y)
plt.plot(c[0], f(c[0]), "ro")
plt.grid()
```

#Al ser cuadratica se tiene un máximo global, en este caso se úbica en x = 6.5

3. $f(x)=5+x+\frac{4}{x}$, para $x>0$

```python
x = sym.var('x', positive = True)
def f(x):
  return 5+x+(4/x)

df = sym.diff(f(x))
d2f = sym.diff(df)
c = sym.solve(df, x)
display(df, d2f, float(c[0]))
x = np.linspace(0, 3, 200)
y = f(x)
plt.plot(x, y)
plt.plot(c[0], f(c[0]), "ro")
plt.grid()
```

#Solo contiene un mínimo local en x = 2

4. $f(x)=\frac{15x}{2}\sqrt{\frac{x}{x-30}}$

```python
x = sym.var('x')
f = (15*x/2*sym.sqrt(x/(x-30)))
df = f.diff()
d2f = df.diff()
c = sym.solve(df, x)
display(df, d2f, c)
f_num = sym.lambdify(x, f, 'numpy')
x_num = np.linspace(-50, 50, 200)
y_num = f_num(x_num)
plt.plot(x_num, y_num)
plt.plot(float(c[0]), f_num(float(c[0])), "ro")
plt.plot(float(c[1]), f_num(float(c[1])), "ro")
plt.grid()
```

#Ahora tenemos un máximo local en x = 0 y otro mínimo local en x = 45

5. $f(x)=6-2x+\sqrt{18-2x^2}$, para $-3\leq x\leq 3$

```python
x = sym.var('x', domain = sym.Interval(-3, 3))
f = (6-2*x+sym.sqrt(18-2*x**2))
df = f.diff()
d2f = df.diff()
c = sym.solve(df, x)
display(df, d2f, float(c[0]))
f_num = sym.lambdify(x, f, 'numpy')
x_num = np.linspace(-3, 3, 200)
y_num = f_num(x_num)
plt.plot(x_num, y_num)
plt.plot(float(c[0]), f_num(float(c[0])), "ro")
plt.grid()
```

#Máximo local en x = -2.4495 y se podría decir que un mínimo global en -3

6. $f(x)=\frac{1}{4}x^4+\frac{1}{3}x^3-x^2$ en el intervalo $[-3,1.7]$

```python
x = sym.var('x', domain = sym.Interval(-3, 1.7))
def f(x):
  return (1/4*x**4)+(1/3*x**3)-(x**2)

df = sym.diff(f(x))
d2f = sym.diff(df)
c = sym.solve(df, x)
display(df, d2f, float(c[0]))
x = np.linspace(-3, 1.7, 200)
y = f(x)
plt.plot(x, y)
plt.plot(c[0], f(c[0]), "ro")
plt.plot(0, 0, "ro")
plt.plot(1, -5/12, "ro")
plt.grid()
```

#Tenemos un mínimo global en x = -2, un máximo local en x = 0 y
#un mínimo local en  x = 1

7. $f(x)=100(1+\cos(x)\sin(x))$, para $x\in[0.\frac{\pi}{2}]$

```python
x = sym.var('x', domain = sym.Interval(0, 1.5707963267948966))
f = 100*(1+sym.cos(x)*sym.sin(x))
df = f.diff()
d2f = df.diff()
c = sym.solve(df, x)
for sol in c:
  if 0 <= sol.evalf() <= np.pi/2:
    rangoc = sol.evalf()
display(df, d2f, rangoc)
f_num = sym.lambdify(x, f, 'numpy')
x_num = np.linspace(0, 2, 200)
y_num = f_num(x_num)
plt.plot(x_num, y_num)
plt.plot(float(rangoc), f_num(float(rangoc)), "ro")
plt.grid()
```
